```python
def remove_punctuation(text):
    final = "".join(u for u in text if u not in ("?", ".", ";", ":",  "!",'"'))
    return final

df = df.dropna(subset=['Summary'])
df['Summary'] = df['Summary'].apply(remove_punctuation)
df['Text'] = df['Text'].apply(remove_punctuation)
```

```python
from sklearn.model_selection import train_test_split

X = df['Text']
Y = df['Sentiment']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(token_pattern=r'\b\w+\b')
x_train_vet = vectorizer.fit_transform(x_train)
x_test_vet = vectorizer.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(penalty='l1', C=0.1, solver='liblinear').fit(x_train_vet, y_train)
```

```python
y_pred = logistic_model.predict(x_test_vet)
```

```python
df_results = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred})
```

```python
df_results.head()
```

```python
df_crosstab = pd.crosstab(df_results.y_pred, df_results.y_test)
df_crosstab
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Acurácia : ", accuracy)
print("Precisão : ", precision)
print("Recall : ", recall)
```

```python
df_crosstab
```

```python
TP = df_crosstab[1][1]
TN = df_crosstab[0][0]
FP = df_crosstab[0][1]
FN = df_crosstab[1][0]
```

```python
accuracy_calc = (TP + TN) / (TP + FP + TN + FN)
print("Acurácia (calculada): ", accuracy_calc)
```

```python
precision_calc = TP / (TP + FP)
msg = "Precisão (calculada) = " + str(precision_calc) + " com Falsos Positivos = " + str(FP)
print(msg)
```

```python
recall_calc = TP / (TP + FN)
msg = "Recall (calculada) = " + str(recall_calc) + " com Falsos Negativos = " + str(FN)
print(msg)
```
